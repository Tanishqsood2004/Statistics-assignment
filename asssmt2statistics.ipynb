(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum). (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers

````python
import random
import statistics
import numpy as np

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Function to calculate the mean of a list
def calculate_mean(lst):
    return sum(lst) / len(lst)

# Function to calculate the median of a list
def calculate_median(lst):
    return statistics.median(lst)

# (ii) Function to compute the mode of a list of integers
def calculate_mode(lst):
    return statistics.mode(lst)

# (iii) Function to calculate the weighted mean of a list
def calculate_weighted_mean(values, weights):
    return np.average(values, weights=weights)

# (iv) Function to calculate the geometric mean of a list of positive numbers
def calculate_geometric_mean(lst):
    return np.exp(np.mean(np.log(lst)))

# (v) Function to calculate the harmonic mean of a list
def calculate_harmonic_mean(lst):
    return len(lst) / sum(1/x for x in lst)

# (vi) Function to calculate the midrange of a list (average of the min and max)
def calculate_midrange(lst):
    return (min(lst) + max(lst)) / 2

# (vii) Function to calculate the trimmed mean (excluding a certain percentage of outliers)
def calculate_trimmed_mean(lst, trim_percentage):
    return statistics.mean(sorted(lst)[int(trim_percentage/100 * len(lst)): -int(trim_percentage/100 * len(lst))])

# Example usage of each function
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weights = [random.random() for _ in range(100)]  # Random weights
weighted_mean = calculate_weighted_mean(int_list, weights)
geometric_mean = calculate_geometric_mean(int_list)
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 10)

# Display results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Weighted Mean:", weighted_mean)
print("Geometric Mean:", geometric_mean)
print("Harmonic Mean:", harmonic_mean)
print("Midrange:", midrange)
print("Trimmed Mean (10% outliers):", trimmed_mean)
````

# Q 3Write a Python class representing a discrete random variable with methods to calculate its expected value and variance

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize the discrete random variable with its values and probabilities.

        :param values: List of possible values of the random variable.
        :param probabilities: List of probabilities corresponding to the values.
        """
        if len(values) != len(probabilities):
            raise ValueError("The length of values and probabilities must be the same.")
        if any(p < 0 or p > 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if sum(probabilities) != 1:
            raise ValueError("The sum of probabilities must be 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the discrete random variable.

        :return: The expected value.
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the discrete random variable.

        :return: The variance.
        """
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

# Create an instance of the DiscreteRandomVariable class
random_variable = DiscreteRandomVariable(values, probabilities)

# Calculate expected value and variance
expected_val = random_variable.expected_value()
variance_val = random_variable.variance()

print("Expected Value:", expected_val)
print("Variance:", variance_val)
````

# Q 4 Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes

````python
import random

class DieSimulation:
    def __init__(self, sides=6, rolls=1000):
        """
        Initialize the die simulation with the number of sides and number of rolls.

        :param sides: Number of sides on the die (default is 6 for a fair six-sided die).
        :param rolls: Number of times to roll the die (default is 1000).
        """
        self.sides = sides
        self.rolls = rolls
        self.outcomes = []

    def roll(self):
        """Simulate the rolling of the die and store the outcome."""
        for _ in range(self.rolls):
            self.outcomes.append(random.randint(1, self.sides))

    def expected_value(self):
        """Calculate the expected value of the die rolls."""
        return sum(self.outcomes) / len(self.outcomes)

    def variance(self):
        """Calculate the variance of the die rolls."""
        mean = self.expected_value()
        return sum((x - mean) ** 2 for x in self.outcomes) / len(self.outcomes)

    def theoretical_expected_value(self):
        """Calculate the theoretical expected value for a fair die."""
        return (self.sides + 1) / 2

    def theoretical_variance(self):
        """Calculate the theoretical variance for a fair die."""
        return (self.sides**2 - 1) / 12

# Example usage
die = DieSimulation(rolls=10000)

# Simulate rolling the die
die.roll()

# Calculate expected value and variance from simulation
simulated_expected_value = die.expected_value()
simulated_variance = die.variance()

# Calculate theoretical expected value and variance
theoretical_expected_value = die.theoretical_expected_value()
theoretical_variance = die.theoretical_variance()

# Display the results
print(f"Simulated Expected Value: {simulated_expected_value:.2f}")
print(f"Simulated Variance: {simulated_variance:.2f}")
print(f"Theoretical Expected Value: {theoretical_expected_value:.2f}")
print(f"Theoretical Variance: {theoretical_variance:.2f}")
````

# Q 5 Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_random_samples(distribution, *params, size=1000):
    """
    Generates random samples from a given probability distribution and calculates their mean and variance.

    :param distribution: The name of the distribution (e.g., 'binomial', 'poisson').
    :param params: Parameters for the distribution (e.g., n, p for binomial; lambda for Poisson).
    :param size: The number of random samples to generate (default is 1000).
    :return: Mean and variance of the generated samples.
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance


# Example usage for binomial distribution
n = 10  # number of trials
p = 0.5  # probability of success
mean_binomial, variance_binomial = generate_random_samples('binomial', n, p, size=10000)

# Example usage for Poisson distribution
lam = 5  # rate (lambda) for the Poisson distribution
mean_poisson, variance_poisson = generate_random_samples('poisson', lam, size=10000)

# Display the results
print(f"Binomial Distribution (n={n}, p={p}):")
print(f"Mean: {mean_binomial:.2f}, Variance: {variance_binomial:.2f}")

print(f"\nPoisson Distribution (lambda={lam}):")
print(f"Mean: {mean_poisson:.2f}, Variance: {variance_poisson:.2f}")
````

# Q 6 Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def generate_gaussian_samples(mean, std_dev, size=1000):
    """
    Generate random samples from a Gaussian (normal) distribution and compute their mean, variance, and standard deviation.

    :param mean: Mean (mu) of the Gaussian distribution.
    :param std_dev: Standard deviation (sigma) of the Gaussian distribution.
    :param size: The number of random samples to generate (default is 1000).
    :return: Mean, variance, and standard deviation of the generated samples.
    """
    # Generate random samples from the normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate the mean, variance, and standard deviation of the samples
    calculated_mean = np.mean(samples)
    variance = np.var(samples)
    std_dev_calculated = np.std(samples)

    return calculated_mean, variance, std_dev_calculated


# Example usage
mean = 0  # Mean (mu) for the Gaussian distribution
std_dev = 1  # Standard deviation (sigma) for the Gaussian distribution
size = 10000  # Number of samples to generate

# Generate samples and compute the mean, variance, and standard deviation
mean_value, variance_value, std_dev_value = generate_gaussian_samples(mean, std_dev, size)

# Display the results
print(f"Generated Gaussian Distribution (mean={mean}, std_dev={std_dev}):")
print(f"Calculated Mean: {mean_value:.4f}")
print(f"Calculated Variance: {variance_value:.4f}")
print(f"Calculated Standard Deviation: {std_dev_value:.4f}")
````

# Q 8 Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.

````python
import numpy as np

def normal_pdf(x, mu, sigma):
    """
    Calculate the Probability Density Function (PDF) for a given normal distribution.

    :param x: The value at which to evaluate the PDF.
    :param mu: The mean (mu) of the normal distribution.
    :param sigma: The standard deviation (sigma) of the normal distribution.
    :return: The PDF value at x.
    """
    # Calculate the PDF using the formula
    pdf_value = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-0.5 * ((x - mu) ** 2) / sigma**2)
    return pdf_value

# Example usage
mu = 0  # Mean of the normal distribution
sigma = 1  # Standard deviation of the normal distribution
x = 0  # Value at which to evaluate the PDF

# Calculate PDF at x = 0
pdf_value = normal_pdf(x, mu, sigma)
print(f"The PDF of the normal distribution at x = {x} is: {pdf_value:.4f}")
````

# Q 9 Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np

def exponential_cdf(x, lambda_):
    """
    Calculate the Cumulative Distribution Function (CDF) for an exponential distribution.

    :param x: The value at which to evaluate the CDF.
    :param lambda_: The rate parameter (lambda) of the exponential distribution.
    :return: The CDF value at x.
    """
    # Calculate the CDF using the formula
    cdf_value = 1 - np.exp(-lambda_ * x)
    return cdf_value

# Example usage
lambda_ = 1  # Rate parameter for the exponential distribution
x = 2  # Value at which to evaluate the CDF

# Calculate CDF at x = 2
cdf_value = exponential_cdf(x, lambda_)
print(f"The CDF of the exponential distribution at x = {x} is: {cdf_value:.4f}")
````

# Q 10 Write a Python function to calculate the probability mass function (PMF) of Poisson distribution

````python
import math

def poisson_pmf(x, lambda_):
    """
    Calculate the Probability Mass Function (PMF) for a Poisson distribution.

    :param x: The number of events (x) at which to evaluate the PMF (x must be a non-negative integer).
    :param lambda_: The rate parameter (lambda) of the Poisson distribution.
    :return: The PMF value at x.
    """
    if x < 0 or not isinstance(x, int):
        raise ValueError("x must be a non-negative integer.")

    # Calculate the PMF using the Poisson formula
    pmf_value = (lambda_**x * math.exp(-lambda_)) / math.factorial(x)
    return pmf_value

# Example usage
lambda_ = 3  # Rate parameter (average number of events)
x = 4  # Number of events

# Calculate PMF at x = 4
pmf_value = poisson_pmf(x, lambda_)
print(f"The PMF of the Poisson distribution at x = {x} with lambda = {lambda_} is: {pmf_value:.4f}")
````

# Q 11 A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare. To generate the data use the following command: ```python import numpy as np # 50 purchases out of 1000 visitors old_layout = np.array([1] * 50 + [0] * 950) # 70 purchases out of 1000 visitors new_layout = np.array([1] * 70 + [0] * 930) ``` Apply z-test to find which layout is successful.

````python
import numpy as np
import scipy.stats as stats

# Data generation
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Calculate the number of successes and sample sizes
x1 = old_layout.sum()  # Number of purchases in old layout
x2 = new_layout.sum()  # Number of purchases in new layout
n1 = len(old_layout)  # Sample size for old layout
n2 = len(new_layout)  # Sample size for new layout

# Proportions of successes in each group
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
p = (x1 + x2) / (n1 + n2)

# Standard error for the difference in proportions
se = np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Z-test statistic
z = (p1 - p2) / se

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # two-tailed test

# Output the results
print(f"Z-test statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion based on the p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject null hypothesis: The new layout has a higher conversion rate.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in conversion rates.")
````

To test if the tutoring service's claim that its program improves students' exam scores is valid, we can use a paired sample z-test. The paired z-test is used when you have two sets of data that are related (such as before and after measurements of the same individuals).

````python
import numpy as np
import scipy.stats as stats

# Data for branch A and branch B customer satisfaction scores
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Function to perform two-sample t-test
def two_sample_t_test(branch_a_scores, branch_b_scores):
    # Calculate the means and standard deviations of both samples
    mean_a = np.mean(branch_a_scores)
    mean_b = np.mean(branch_b_scores)
    std_a = np.std(branch_a_scores, ddof=1)
    std_b = np.std(branch_b_scores, ddof=1)

    n_a = len(branch_a_scores)
    n_b = len(branch_b_scores)

    # Calculate the t-statistic
    t_stat = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

    # Calculate the degrees of freedom
    df = ((std_a**2 / n_a + std_b**2 / n_b)**2) / (
        ((std_a**2 / n_a)**2) / (n_a - 1) + ((std_b**2 / n_b)**2) / (n_b - 1)
    )

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

    # Output the results
    print(f"t-statistic: {t_stat:.4f}")
    print(f"Degrees of freedom: {df:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Conclusion based on the p-value
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject null hypothesis: There is a significant difference in customer satisfaction between the two branches.")
    else:
        print("Fail to reject null hypothesis: There is no significant difference in customer satisfaction between the two branches.")

# Perform two-sample t-test analysis
two_sample_t_test(branch_a_scores, branch_b_scores)
````

# Q 21 A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences. Use the below code to generate data: ```python np.random.seed(0) age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30) voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

To perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences, we can follow these steps:

Create a contingency table: A contingency table summarizes the frequency of occurrences of combinations of two categorical variablesin this case, age_groups and voter_preferences.

Chi-Square Test: The Chi-Square test of independence checks whether there is a significant association between two categorical variables. It compares the observed frequencies in the contingency table with the expected frequencies under the assumption of no association.

Hypotheses:

Null hypothesis (

0
H
0

 ): There is no association between age groups and voter preferences (i.e., the variables are independent).
Alternative hypothesis (

1
H
1

 ): There is an association between age groups and voter preferences (i.e., the variables are not independent).
Chi-Square Test Calculation: We will calculate the Chi-Square statistic and the p-value to determine if the result is statistically significant.

````python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Generate synthetic data for age groups and voter preferences
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)  # Fixed size for simplicity
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame to display the data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table (cross-tabulation of age groups and voter preferences)
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Output the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_value)

# Conclusion based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject null hypothesis: There is no significant association between age groups and voter preferences.")
````

# Q 24 A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores. Use the following data: ```python # Sample data: Customer satisfaction scores for each product version standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87] ``` premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93] deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

To determine if there is a significant difference in customer satisfaction scores among the three product versions (Standard, Premium, and Deluxe), we can perform an ANOVA (Analysis of Variance) test. This statistical test allows us to compare the means of three or more independent groups to see if at least one group mean is significantly different from the others.

Hypotheses:
Null Hypothesis (

0
H
0

 ): The mean customer satisfaction scores are the same for all three product versions (i.e., no significant difference).
Alternative Hypothesis (

1
H
1

 ): At least one product version has a different mean customer satisfaction score (i.e., there is a significant difference).
Steps:
Data: We have the customer satisfaction scores for each product version:
standard_scores
premium_scores
deluxe_scores
ANOVA Test: We use the f_oneway() function from scipy.stats to perform the one-way ANOVA test.

````python
import numpy as np
from scipy import stats

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

# Output the results
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)

# Conclusion based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nReject null hypothesis: There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("\nFail to reject null hypothesis: There is no significant difference in customer satisfaction scores among the product versions.")
````
